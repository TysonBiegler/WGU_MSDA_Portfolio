# Handling Outliers

### Population

```python
plt.figure(figsize=(8, 6))
plt.boxplot(churn['Population'])
plt.title('Boxplot of Population')
plt.xlabel('Population')
plt.show()
```

```python
# Summary statistics
Q1 = churn['Population'].quantile(0.25)
Q3 = churn['Population'].quantile(0.75)

# Interquartile Range (IQR)
IQR = Q3 - Q1

# outliers
outlier_threshold_upper = Q3 + 1.5 * IQR
outlier_threshold_lower = Q1 - 1.5 * IQR

print("Outliers begin above:", outlier_threshold_upper)
```

```python
# Population
churn['Population'] = np.where(churn['Population'] > 31813, churn['Population'], churn['Population']) # Retaining outliers
```

### Email

```python
plt.figure(figsize=(8, 6))
plt.boxplot(churn['Email'].dropna())
plt.title('Boxplot of Email')
plt.xlabel('Email')
plt.show()
```

```python
# Summary statistics
Q1 = churn['Email'].quantile(0.25)
Q3 = churn['Email'].quantile(0.75)

# Interquartile Range (IQR)
IQR = Q3 - Q1

# outliers
outlier_threshold_upper = Q3 + 1.5 * IQR
outlier_threshold_lower = Q1 - 1.5 * IQR

print("Outliers begin above:", outlier_threshold_upper)
print("Outliers begin below:", outlier_threshold_lower)
```

```python
# Email
churn['Email'] = np.where((churn['Email'] < 4) | (churn['Email'] > 20), np.nan, churn['Email'])
churn['Email'].fillna(churn['Email'].mean(), inplace=True)
```

### Other Data Quality Issues

```python
# Removing white space and ensuring that all Zips are 5 digets
churn['Zip'] = churn['Zip'].astype(str).str.zfill(5).str.strip()
```

```python
# making Gender match data dictionary
churn['Gender'] = churn['Gender'].replace({"Prefer not to answer": "Nonbinary"}).astype('category')
print(churn.Gender.unique())
```

```python
# converting columns to categorical
categorical_columns = ['Contract', 'InternetService', 'Area', 'Marital', 'State', 'PaymentMethod']
churn[categorical_columns] = churn[categorical_columns].astype('category')
```

```python
# Converting columns to Numeric
numeric_columns = ['Tenure', 'MonthlyCharge', 'Bandwidth_GB_Year', 'Outage_sec_perweek']
churn[numeric_columns] = churn[numeric_columns].apply(pd.to_numeric)
```

```python
# Converting to Boolean Data Type
bool_columns = [
    'Churn', 'Techie', 'Port_modem', 'Tablet', 'Phone', 'Multiple',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies', 'PaperlessBilling'
]
```

```python
for col in bool_columns:
    churn[col] = churn[col].str.strip().apply(lambda x: True if x == 'Yes' else False)
```

```python
# Updating Payment Method to Match Data Dictionary
churn['PaymentMethod'] = churn['PaymentMethod'].str.lower()
churn['PaymentMethod'] = churn['PaymentMethod'].replace({
    "bank transfer(automatic)": "bank (automatic bank transfer)"
})
churn['PaymentMethod'] = churn['PaymentMethod'].astype('category')
```

```python
# Rounding Numeric Columns
churn['Tenure'] = churn['Tenure'].round(2)
churn['MonthlyCharge'] = churn['MonthlyCharge'].round(2)
churn['Bandwidth_GB_Year'] = churn['Bandwidth_GB_Year'].round(0)
churn['Income'] = churn['Income'].round(2)
churn['Outage_sec_perweek'] = churn['Outage_sec_perweek'].round(0)
```

```python
# Renaming Survey Response Columns
churn = churn.rename(columns={
    churn.columns[42]: "Timely_response",
    churn.columns[43]: "Timely_fixes",
    churn.columns[44]: "Timely_replacements",
    churn.columns[45]: "Reliability",
    churn.columns[46]: "Options",
    churn.columns[47]: "Respectful",
    churn.columns[48]: "Courteous",
    churn.columns[49]: "Active_listening"
})
print(churn.columns)
```

```python
# Rename columns to consistent naming conventions
churn = churn.rename(columns={
    'InternetService': 'Internet_service',
    'OnlineSecurity': 'Online_security',
    'TechSupport': 'Tech_support',
    'PaperlessBilling': 'Paperless_billing',
    'CaseOrder': 'Index',
    'OnlineBackup': 'Online_backup',
    'StreamingTV': 'Streaming_TV',
    'PaymentMethod': 'Payment_method',
    'Customer_id': 'Customer_ID',
    'DeviceProtection': 'Device_protection',
    'StreamingMovies': 'Streaming_movies',
    'MonthlyCharge': 'Monthly_charge',
    'Bandwidth_GB_Year': 'Bandwidth_GB_year',
    'Outage_sec_perweek': 'Outage_sec_per_week'
})
```

```python
import statsmodels.api as sm

# Define the independent variable (Bandwidth_GB_year) and dependent variable (Monthly_charge)
X = churn['Bandwidth_GB_year']
y = churn['Monthly_charge']

# Add a constant (intercept) to the independent variable
X = sm.add_constant(X)

# Create the model
model = sm.OLS(y, X)

# Fit the model and get the summary
results = model.fit()
print(results.summary())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot with a regression line
sns.lmplot(x='Bandwidth_GB_year', y='Monthly_charge', data=churn, hue='Churn', aspect=1.5, height=6, logistic=False)

# Set plot title and labels
plt.title('Monthly Charge vs Bandwidth_GB_year with Regression Line')
plt.xlabel('Bandwidth_GB_year')
plt.ylabel('Monthly Charge')

# Show the plot
plt.show()
```

```python
# List of explainer variables (independent variables)
explainer_vars = ['Tenure', 'Income', 'Children', 'Age', 'Population', 'Contacts', 
                  'Yearly_equip_failure', 'Outage_sec_per_week', 'Email', 'Monthly_charge']

# Select the independent variables
X = churn[explainer_vars]  

# The dependent variable (MonthlyCharge)
y = churn['Monthly_charge']

# Add a constant (intercept) to the independent variables
X = sm.add_constant(X)

# Create the OLS model
model = sm.OLS(y, X)

# Fit the model
results = model.fit()

# Print the summary of the regression model
print(results.summary())
```
