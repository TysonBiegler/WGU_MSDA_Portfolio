```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import figure
import os

os.getcwd()  # This prints the current working directory

df = pd.read_csv("teleco_time_series.csv", index_col='Day')
df=df.dropna()
print('Shape of data: ', df.shape)
df.head()
df.tail()
```

```python
df['Revenue'].plot(figsize=(12,5))
```

```python
from statsmodels.tsa.stattools import adfuller

def ad_test(dataset):
    dftest=adfuller(dataset, autolag = 'AIC')
    print("1. ADF: ", dftest[0])
    print("2. P-value: ", dftest[1])
    print("3. Num of Lags: ", dftest[2])
    print("4. Num of Obs Used for ADF Regression and Critical Values Calc: ", dftest[3])
    print("5. Critical Values: ", dftest[4])
    for key, val in dftest[4].items():
        print("\t", key,": ",val)
        
ad_test(df['Revenue'])

print("Done with differencing now that I have the 5 goals completed")
```

```python
from statsmodels.tsa.arima.model import ARIMA
print(df.shape)
train=df.iloc[:-30] #assign all other data to train
test=df.iloc[-30] #get the last 30 days of the dataset for test
print(train.shape, test.shape)
```

```python
model=ARIMA(train["Revenue"],order=(1,1,0))
model=model.fit()
model.summary()
```

```python
#making predictions
start=len(train)
end=len(train)+len(test)-1
pred=model.predict(start=start, end=end, type='levels')
print(pred)
#set an index for plotting
pred.index=df.index[start:end+1]
#print(pred)
```

```python
#pred.index=df.index[start:end+1]
pred.plot(legend=True)
test['Revenue'].plot(legend=True)
```

```python
test['Revenue'].mean()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rmse = sqrt(mean_squared_error(pred,test['Revenue']))
print(rmse)
```
